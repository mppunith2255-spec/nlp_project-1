```python
import pandas as pd #for uploading the data
import numpy as np #
import re #data cleaning  ---- re -- regular expression
import nltk
import matplotlib.pyplot as plt
from nltk.corpus import stopwords  # corpus -- collection of documents, stop words -- conjuctions, with,an, and, etc,.... #to remove unncesary words
from nltk.tokenize import word_tokenize  # tokenizer
from nltk.sentiment import SentimentIntensityAnalyzer
from collections import Counter
from wordcloud import WordCloud # to visualize the words
```

```python
#Download required NLTK files
nltk.download('punkt')   #punction
nltk.download('stopwords')  #a, an, the, is, am, are, in, on, at, and, of, to, for
nltk.download('vader_lexicon')  #Valence aware dictionary and sentiment reasoner, rule based sentiment analysisS
nltk.download('punkt_tab')
```

```python
data = {
    "Review": [
        "The product is very very good",
        "The product is broken and not functioning",
        "Late delivery and bad customer supprt",
        "Fast shipping and good packing",
        "Excellent service and working neatly"
    ]
}


#create DataFrame
df = pd.DataFrame(data)
```

```python
#Display first few rows
print("Customer Review Dtatset")
print(df.head())
```

```python
stop_words = set(stopwords.words('english')) #set to remove the duplicate values in the dataset
def clean_text(text):

   #convert to lowercase
   text = text.lower()

   #remove special characters and punctuation
   text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  #sub---substitue

   #Tokenization
   words = word_tokenize(text)   #word tokenize

   #remove stop words
   words = [word for word in words if word not in stop_words]  #List comprehension
   #LIST COMPREHENSION---IF THE WORD IS THERE OK IF NOT ADD THOSE WORDS IN NEXT STEP

   #Join words again
   cleaned_text = " ".join(words)

   return cleaned_text

   #Apply cleaning function
df['cleaned_Review'] = df['Review'].apply(clean_text)

print("\n Cleaned Review Dataset") #printing
print(df[['Review', 'cleaned_Review']])
```
